```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt

cubes = cv2.imread('cubes.png')

#converting from OpenCV to PIL
cubes = cv2.cvtColor(cubes, cv2.COLOR_BGR2RGB)

# Operação reversa:
#cubes = np.asarray(cubes_pil)
```

```python
#show PIL image
imgplot = plt.imshow(cubes)
plt.title('Cubes'), plt.xticks([]), plt.yticks([])
```

```python
cubesLAB = Image.fromarray(cubes)

plt.imshow(cubesLAB)
plt.title('Cubes'), plt.xticks([]), plt.yticks([])
```

```python
#l,a,b = cubesLAB.split()
import numpy as np
def channel(img, n):
    """Isolate the nth channel from the image.

       n = 0: red, 1: green, 2: blue
    """
    a = np.array(img)
    a[:,:,(n!=0, n!=1, n!=2)] *= 0
    return Image.fromarray(a)

l = channel(cubesLAB,0)
a = channel(cubesLAB,1)
b = channel(cubesLAB,2)

plt.subplot(2,2,1),plt.imshow(l)
plt.title('L'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2),plt.imshow(a)
plt.title('A'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(b)
plt.title('B'), plt.xticks([]), plt.yticks([])
```

```python
cubesYcrcb = cv2.cvtColor(cubes, cv2.COLOR_RGB2YCrCb)

plt.imshow(cubesYcrcb)
plt.title('Cubes'), plt.xticks([]), plt.yticks([])
```

```python
cubesYcrcbPIL = Image.fromarray(cubesYcrcb)
y = channel(cubesYcrcbPIL,0)
cr = channel(cubesYcrcbPIL,1)
cb = channel(cubesYcrcbPIL,2)

plt.subplot(2,2,1),plt.imshow(y)
plt.title('Y'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2),plt.imshow(cr)
plt.title('Cr'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(cb)
plt.title('Cb'), plt.xticks([]), plt.yticks([])
```

```python
cubesHSV = cv2.cvtColor(cubes, cv2.COLOR_RGB2HSV)

imgplot = plt.imshow(cubesHSV)
plt.title('Cubes'), plt.xticks([]), plt.yticks([])
```

```python
cubesHSVPIL = Image.fromarray(cubesHSV)
h = channel(cubesHSVPIL,0)
s = channel(cubesHSVPIL,1)
v = channel(cubesHSVPIL,2)

plt.subplot(2,2,1),plt.imshow(h)
plt.title('H'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,2),plt.imshow(s)
plt.title('S'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(v)
plt.title('V'), plt.xticks([]), plt.yticks([])
```
